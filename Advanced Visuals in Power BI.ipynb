### Advanced Visuals in Power BI

**1. What is the difference between a Basic Filter, Advanced Filter, and a Slicer in Power BI?**
- Difference between the three in Power BI:

  - **Basic Filter:** Simple filter applied to a visual, page, or report. Lets us select values from a list (e.g., show only “East” region).
  
  - **Advanced Filter:** Allows more complex conditions using logical operators (e.g., greater than, less than, contains).

  - **Slicer:** A visual filter that users can interact with directly on the report. Makes filtering dynamic and user-friendly for dashboards

<br>

**4. What is a Visual-Level Filter, and how does it differ from Page-Level and Report-Level Filters?**
- A Visual-Level Filter is a filter that applies only to a single visual in a report. It controls what data appears inside that one visual, without affecting anything else on the page or report.

- **Difference between Visual-Level, Page-Level and Report-Level Filters:**
  - Visual-level filters *apply to only one visual Like chart, table, etc.*
  - Page-level filters *apply to all visuals on a single page.*
  - Report-level filters *apply to all pages and visuals in the entire report.*

<br>

**6. What is the role of Drill-Down in Power BI? How does it improve data exploration?**
- Drill-down in Power BI allows users to click through hierarchical data (Like: Year, Month, & Day) to see more detailed information within a visual, making data exploration interactive and faster while keeping reports clean.

<br>

**10. Explain the difference between Drill-Down and Drill-Through in Power BI using a real-world example.**
- **Drill-Down:** Deeper into Hierarchy
  - We stay on the same visual and move down through hierarchical levels (Like: Year → Month).
  - Example: Clicking on 2025 in a bar chart to view sales for individual months within that year.

- **Drill-Through:** Detailed Deep Dive
  - We navigate to a different report page that is automatically filtered for a selected item.
  - Example: Right-clicking a specific Product Name and opening a page that displays its inventory, supplier details, and customer reviews.

<br>

**13. What is the “Keep All Filters” option in Drill-Through? How does it affect user interaction**
- In Power BI, Keep All Filters is a setting on the drill-through destination page that controls whether all existing filters from the source page are carried over when a user drills through.

- **How it Affects User Interaction:**
  - **Enabled:** The drill-through page keeps the full context of the user’s selections, giving a more precise and consistent analysis.
  - **Disabled:** Only the drill-through field (e.g., Product or Region) is applied, allowing users to see a broader or cleaner view of the data.

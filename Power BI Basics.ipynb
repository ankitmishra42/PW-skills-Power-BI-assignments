## Power BI Basics

**Q1: What is Power BI and why is it used in businesses?**
- Power BI is a business intelligence and data visualization tool by Microsoft. It is used in businesses to connect data from multiple sources, clean and model it, and create interactive dashboards and reports.

<br>

**Q2: Name and explain the three main components of Power BI.**
- Three main components of Power BI:
    - **Power BI Desktop:** Used to connect, transform, and model data and to create reports and dashboards on a local system.
    - **Power BI Service:** A cloud-based platform used to publish, share, and collaborate on reports and dashboards with others.
    - **Power BI Mobile:** A mobile app that allows users to view and interact with reports and dashboards anytime, anywhere on smartphones and tablets.

<br>

**Q3: Explain the Power BI workflow.**
- **Connect to Data Sources:** First, we import data from databases, files, APIs, or cloud services.
- **Transform Data:** Then, we clean and shape the data using Power Query.
- **Data Modeling:** After that, we create relationships, measures, and calculated columns using DAX.
- **Build Reports:** Next, we design interactive visualizations and reports in Power BI Desktop.
- **Publish & Share:** Finally, we publish the reports to Power BI Service and share dashboards with users.

**Q6: Define Data View, Report View, and Model View. Explain the purpose of each view.**
- **Data View:** Used to view and verify data in table format.
- **Report View:** Used to create visualizations and dashboards.
- **Model View:** Used to manage relationships between tables.

<br>

**Q7: Discuss the different data sources that Power BI supports.**
- Different data sources that Power BI supports:
  - **Files:** Excel, CSV, XML, JSON, PDF
  - **Databases:** SQL Server, MySQL, PostgreSQL, Oracle
  - **Cloud services:** Azure SQL, Azure Blob Storage, Google Analytics
  - **Online services:** SharePoint, OneDrive, Dynamics 365
  - **Other sources:** Web data, APIs, OData feeds

**Q8: Split Owner Name to create two new columns as First Name and Last Name**
- Steps to split Owner Name into First Name and Last Name in Power Query:
  - Open Power BI Desktop and click on Transform Data.
  - Select the Owner Name column.
  - Go to the Transform tab and click Split Column.
  - Choose By Delimiter.
  - Select Space as the delimiter.
  - Split at the left-most delimiter.
  - Rename the new columns as First Name and Last Name.

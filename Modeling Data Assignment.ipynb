**Theoretical Questions**<br>
**Q1: What is Data Loading in Power BI and why is it considered the first step of analysis?**
- Data Loading in Power BI means importing data from different sources (Excel, CSV, databases, APIs, cloud services, etc.) into Power BI so it can be analyzed and visualized.

- It is considered the first step of analysis because:
  - Without loading data, there is nothing to analyze or visualize.
  - It helps Power BI understand the structure of the data (columns, data types, rows).
  - It allows us to check whether the data is complete, relevant, and usable.
  - All next steps like data cleaning (Power Query), modeling (relationships, DAX), and report building depend on the data that is loaded first.

<br>

**Q2: Explain the difference between “Load” and “Transform Data” in Power BI.**
- Load:
  - Load means directly importing the data into Power BI as it is.
  - The data is brought into the Power BI data model without any changes.
  - It is used when the data is already clean and well-structured.
  - Faster, but gives no control over cleaning or shaping the data.

- Transform Data:
  - Transform Data opens Power Query Editor.
  - It is used to clean, modify, and shape the data before loading it.
  - e.g.- removing duplicates, changing data types, splitting columns, filtering rows.
  - Ensures data quality and better analysis.

<br>

**Q3: What is a Fact Table and a Dimension Table? Give examples from the dataset.**
- **Fact Table**
  - A Fact Table stores quantitative, numerical data that we analyze and aggregate.
  - It usually contains measures like counts, totals, or sums.
  - It also has foreign keys that connect to dimension tables.
  - Example from the Corona_Virus_2020 dataset:
    - Confirmed_Cases
    - Recovered_Cases
    - Deaths
    - Active_Cases
    - Tests_Conducted
    - Hospitalized

- These columns represent measurable values, so the main COVID records table acts as a fact table.

- **Dimension Table**
  - A Dimension Table stores descriptive information used to slice and filter data.
  - It answers questions like when, where, and who.
  - Example dimensions from the dataset:
    - Date
    - Country
    - State
    - Vaccination_Status

<br>

**Q4: Why is Star Schema preferred over Snowflake Schema in Power BI?**
- **Star Schema is preferred because:**
  - It has a simple structure: one fact table connected directly to multiple dimension tables.
  - Better performance: fewer joins mean faster query execution.
  - Easy to understand for both developers and business users.
  - Simpler DAX calculations due to fewer relationships.
  - Works best with Power BI’s in-memory (VertiPaq) engine.

- **Snowflake Schema drawbacks in Power BI:**
  - More tables and relationships.
  - Complex joins reduce performance.
  - Harder to write and maintain DAX measures.

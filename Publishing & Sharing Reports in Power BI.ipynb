### Publishing & Sharing Reports in Power BI

### Easy Level

**Q1. Explain why publishing and sharing reports is considered more important than just building dashboards in Power BI Desktop.**
- Publishing and sharing reports matters more because insights only create value when others can access and use them. Power BI Desktop dashboards are limited to one user, while published reports allow collaboration, real-time data refresh, secure access, and decision-making across the organization.

<br>

**Q2. Define Power BI Service and explain its role in comparison to Power BI Desktop in the reporting lifecycle.**

- Power BI Service is a cloud-based platform used to publish, share, and manage Power BI reports and dashboards.

- **Role of Power BI Service in the reporting lifecycle:** Power BI Desktop is used to create and design reports (data modeling, visuals, calculations), while Power BI Service is used to deploy, refresh, secure, and share those reports with users so insights can be accessed and acted upon across the organization.

<br>

**Q3. What does publishing a report mean in Power BI? List any two outcomes that occur after a report is published.**
- Publishing a report in Power BI means uploading a report created in Power BI Desktop to the Power BI Service (cloud) so it can be accessed and shared online.

- **Two outcomes after publishing:**
  - The report and its dataset become available in the Power BI Service workspace.
  - The report can be shared with users, viewed in a browser or mobile app, and set up for scheduled data refresh.

<br>

### Medium Level

**Q4. List and explain the steps involved in publishing a report from Power BI Desktop to Power BI Service in the correct sequence.**
- Steps to publish a report in Power BI:
  - **Save:** Ensure .pbix file is saved locally.
  - **Sign In:** Click Sign In (top-right) using Power BI.
  - **Publish:** On the Home tab, click the Publish button.
  - **Select Workspace:** Choose destination (e.g., "My Workspace" for private drafts or a specific Team Workspace for collaboration).
  - **Confirm Success:** Click the link in the "Success!" dialog to open the report in browser.
  - **Set Refresh:** In the Power BI Service, go to the Dataset settings to configure Data Gateway and schedule automatic updates.

<br>

**Q5. Differentiate between Report and Dashboard in Power BI based on structure and usage.**
- **Report:**
A report is a multi-page, interactive collection of visuals built in Power BI Desktop. It is used for detailed analysis and allows users to filter, drill down, and explore data.

- **Dashboard:**
A dashboard is a single-page view created in Power BI Service by pinning visuals from reports. It is used for high-level monitoring of key metrics and provides a quick overview.

<br>

**Q6. Explain the difference between the following three sharing methods in Power BI Service: Direct Report Sharing, Workspace Access, Power BI Apps, Also mention one use case for each.**

- **Direct Report Sharing**
  - Reports are shared directly with specific users via email or link, with view or edit permissions.
  - **Use case:** Sharing a single report with a manager for quick review.

- **Workspace Access**
  - Users are added to a workspace with roles (Viewer, Contributor, Member, Admin) to collaborate on multiple reports and datasets.
  - **Use case:** A BI team working together to develop and maintain reports.

- **Power BI Apps**
  - A packaged, read-only collection of dashboards and reports distributed to many users in a controlled way.
  - **Use case:** Publishing finalized reports to a large group like sales or executives.

### Hard Level

**Q8. Why are Power BI Apps considered more secure and scalable than direct sharing? Explain this from a governance and enterprise usage perspective.**
- Power BI Apps are considered more secure and scalable than direct sharing because they support centralized control, consistency, and enterprise governance.

- **From a governance and enterprise perspective:**
  - **Centralized content management:** Apps are published from a workspace by controlled owners, ensuring users see only approved and standardized reports.
  - **Controlled access:** App permissions are managed at one place, reducing the risk of accidental oversharing compared to sharing individual reports.
  - **Read-only distribution:** Most users consume content without edit rights, protecting datasets and report integrity.
  - **Scalability:** Apps can be distributed to large user groups like departments or entire organizations without managing individual shares.
  - **Version control:** Updates to reports automatically reflect in the app, ensuring everyone uses the latest version.

<br>

**Q9. Explain Row-Level Security (RLS) in Power BI. How does RLS ensure that different users see different data using the same report?**
- Row-Level Security (RLS) in Power BI is a feature that restricts data access for users based on their roles, ensuring each user sees only the data they are authorized to view.

- **How it works:**
  - In Power BI Desktop, we define roles with DAX filters on tables.
  - When the report is published to Power BI Service, users are assigned to these roles.
  - Power BI applies the filters automatically, so different users see different subsets of data in the same report, without creating separate reports.

- **Example:**
  - A sales manager in the East region will see only East region data, while a West region manager sees only West region data, all from the same report.

<br>

**Q10.Assume an organization shares reports without using Apps, workspace roles, or RLS. Discuss any three risks or issues that may arise related to security, governance, or data misuse, and explain how Power BI best practices help prevent them**
- **If an organization shares reports without using Apps, workspace roles, or RLS, several risks can arise:**

  - **Unauthorized access:** Anyone with a shared link could view sensitive data.
    - Best practice - Use workspace roles and Power BI Apps to control who can access reports.

  - **Data inconsistency or version issues:** Users may work with outdated copies of reports, leading to wrong  decisions.
    - Best practice - Publish reports via Apps so all users see the latest, standardized version.

  - **Data misuse or leakage:** Without RLS, users may see data not meant for them (e.g., other regions or departments).
    - Best practice - Implement Row Level Security to restrict data visibility per user role.
